### Introduction and Approach

The study intends to identify whether a particular pass is completed or not, on the basis of a number of features. The dataset includes instances of passes from a season of football, along with their values for all the features and identification of them as either completed or not.

The below analysis intends to identify the features which are most indicative of passing completion probability. For this purpose, the below techniques will be employed:

Detailed EDA on the identified features
Unsupervised learning method: K Means Clustering
Supervised learning methods: (Logistic Regression & XGBoost)
Correlation Matrix
EDA on the identified features

The dataset will be split into train & test  sets. The models will be trained on train set and evaluated on test set. The final model will give the results for test set.

The end result will be used for player performance analysis, answering questions such as:
   - Does pass difficulty change in different areas of the pitch
   - Deeper insights into player 263
   - Do groups of players have a passing style?
   - Who are the top passers?

Model sability in a business and sports setting are the most important aspects for the model, rather than performance.

```python
df.head()
```

```python
def result(row):
    if ((row['home_score']) > row['away_score']):
        val = 'H'
    elif ((row['home_score']) < row['away_score']):
        val = 'A'
    else:
        val = 'D'
    return val
```

```python
def own_result(row):
    if ((row['team_location'] == 'home') & (row['game_result'] == 'H')):
        val = 'Win'
    elif ((row['team_location'] == 'away') & (row['game_result'] == 'A')):
        val = 'Win'
    elif ((row['team_location'] == 'home') & (row['game_result'] == 'A')):
        val = 'Loss'
    elif ((row['team_location'] == 'away') & (row['game_result'] == 'H')):
        val = 'Loss'
    else:
        val = 'Draw'
    return val
```

```python
df['game_result'] = df.apply(result, axis=1)
df['own_team_result'] = df.apply(own_result, axis = 1)
```

Passes in general start in mid-field and end near the edges of the field, with most failed passes started in the teams box, or as corners. THey are generally stopped at the opponents box. For this reason I divide up the field into 9, to attempt to gain a deeper insights into the parts of the pitch.

```python
def origin_ninetile(row):
    if ((row['x'] <= 35 ) & (row['y'] <= (68/3))):
        val = 1
    elif ((row['x'] <= 35 ) & ((2*(68/3)) >= row['y'] > (68/3))):
        val = 2
    elif ((row['x'] <= 35 ) & (row['y'] > (2*(68/3)))):
        val = 3
        
        
    elif ((35 < row['x'] <= 70) & (row['y'] <= (68/3))):
        val = 4
    elif ((35 < row['x'] <= 70) & ((2*(68/3)) >= row['y'] > (68/3))):
        val = 5
    elif ((35 < row['x'] <= 70) & (row['y'] > (2*(68/3)))):
        val = 6
        
        
    elif ((row['x'] > 70 ) & (row['y'] <= (68/3))):
        val = 7
    elif ((row['x'] > 70 ) & ((2*(68/3)) >= row['y'] > (68/3))):
        val = 8
    elif ((row['x'] > 70 ) & (row['y'] > (2*(68/3)))):
        val = 9
    return val
```

```python
def end_ninetile(row):
    if ((row['endx'] <= 35 ) & (row['endy'] <= (68/3))):
        val = 1
    elif ((row['endx'] <= 35 ) & ((2*(68/3)) >= row['endy'] > (68/3))):
        val = 2
    elif ((row['endx'] <= 35 ) & (row['endy'] > (2*(68/3)))):
        val = 3
        
        
    elif ((35 < row['endx'] <= 70) & (row['endy'] <= (68/3))):
        val = 4
    elif ((35 < row['endx'] <= 70) & ((2*(68/3)) >= row['endy'] > (68/3))):
        val = 5
    elif ((35 < row['endx'] <= 70) & (row['endy'] > (2*(68/3)))):
        val = 6
        
        
    elif ((row['endx'] > 70 ) & (row['endy'] <= (68/3))):
        val = 7
    elif ((row['endx'] > 70 ) & ((2*(68/3)) >= row['endy'] > (68/3))):
        val = 8
    elif ((row['endx'] > 70 ) & (row['endy'] > (2*(68/3)))):
        val = 9
    return val
```

```python
df['pass_orgin'] = df.apply(origin_ninetile, axis=1)
df['pass_end'] = df.apply(end_ninetile, axis=1)
```

# Clustering Pass Types

```python
df2 = df
```

```python
df2['pass_orgin'] = pd.Categorical(df2.pass_orgin)
df2['pass_end'] = pd.Categorical(df2.pass_end)
df2['half'] = pd.Categorical(df2.half)


df2 = pd.get_dummies(df2)
df2 = df2.drop(columns = ['team_id' ,'team_id', 'game_id','player_id',
                          'home_id','away_id', 'minute', 'second'])
```

```python
train_x, test_x, train_y, test_y = train_test_split(df2.drop(columns = 'outcome'), df2['outcome'], random_state = 42)


non_dummies = ['x', 'y', 'endx', 'endy','pass_distance']
scaler = StandardScaler()

train_x[non_dummies] = scaler.fit_transform(train_x[non_dummies])
test_x[non_dummies] = scaler.transform(test_x[non_dummies])
```

Here I scale the variables which are not dummies using standard scalar, and I drop all columns which are either ID's or time identifiers which I do not believe will help cluster pass type. The scalar is fitted on the training data only to prevent leakage.

To find optimal K for K means, I look at both the silhouette scores, and use the elbow method. However, given the large size of the dataset I use a random sample of 10% of the training data to calculate this.

```python
sampled = train_x.sample(25000)
range_n_clusters = list(range(3, 50))
for n_clusters in range_n_clusters:
    clusterer = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = clusterer.fit_predict(sampled)

    # The silhouette_score gives the average value for all the samples.
    # This gives a perspective into the density and separation of the formed
    # clusters
    silhouette_avg = silhouette_score(sampled, cluster_labels)
    print("For n_clusters =", n_clusters,
          "The average silhouette_score is :", silhouette_avg)
```

```python
wcss = []
for i in range(1, 30):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(sampled)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 30), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 4, random_state = 42).fit(train_x)
cluster_labels = kmeans.predict(train_x)

train_x['total_pass_cluster'] = cluster_labels

centroids_1 = pd.DataFrame(data = kmeans.cluster_centers_[:,[2, 3, 4, 5, 14]], columns = ['x', 'y', 'endx', 'endy', 'pass_distance'])
centroids = scaler.inverse_transform(centroids_1[non_dummies].values)
centroids_orig = pd.DataFrame(data = centroids, columns = ['x', 'y', 'endx', 'endy', 'pass_distance'])
```

```python
pitch = Pitch(pitch_type='statsbomb', orientation='vertical',
              pitch_color='#22312b', line_color='#c7d5cc', figsize=(16, 11),
              constrained_layout=False, tight_layout=False)
fig, ax = pitch.draw()
lc1 = pitch.lines(centroids_orig['x'], centroids_orig['y'],
                  centroids_orig['endx'], centroids_orig['endy'],
                  lw=5, transparent=False, comet=True, ax=ax)


labels = centroids_orig.index

# Set the title
ax.set_title(f'Total Pass Clusters ', fontsize=20, color = 'grey' )

annot_dy = -0.5
annot_dx = -0.5
for i, txt in enumerate(centroids_orig.index):
    ax.annotate(txt, (centroids_orig.iloc[txt]['y'] + annot_dx, centroids_orig.iloc[txt]['x']+ annot_dy), ha='center', color = 'white', fontsize = 15 )

plt.show()
```

Here, although 4 clusters provided the best silhouette score it is quite difficult to interpret anything from them visually. I decide to increase the number of K in K means to 22, with the hope it will provide roughly 2 passes per player, and give a more vizual intepretation of the clusters.

```python
train_x = train_x.drop(columns = 'total_pass_cluster')
# test_x = test_x.drop(columns = 'total_pass_cluster')
```

```python
kmeans = KMeans(n_clusters = 22, random_state = 42).fit(train_x)

cluster_labels_train = kmeans.predict(train_x)
cluster_labels_test = kmeans.predict(test_x)

train_x['total_pass_cluster'] = cluster_labels_train
test_x['total_pass_cluster'] = cluster_labels_test

centroids_1 = pd.DataFrame(data = kmeans.cluster_centers_[:,[2, 3, 4, 5, 14]], columns = ['x', 'y', 'endx', 'endy', 'pass_distance'])
centroids = scaler.inverse_transform(centroids_1[non_dummies].values)
centroids_orig = pd.DataFrame(data = centroids, columns = ['x', 'y', 'endx', 'endy', 'pass_distance'])

pitch = Pitch(pitch_type='statsbomb', orientation='vertical',
              pitch_color='#22312b', line_color='#c7d5cc', figsize=(16, 11),
              constrained_layout=False, tight_layout=False)
fig, ax = pitch.draw()
lc1 = pitch.lines(centroids_orig['x'], centroids_orig['y'],
                  centroids_orig['endx'], centroids_orig['endy'],
                  lw=5, transparent=False, comet=True, ax=ax)


labels = centroids_orig.index

# Set the title
ax.set_title(f'Total Pass Clusters ', fontsize=20, color = 'grey' )

annot_dy = -0.5
annot_dx = -0.5
for i, txt in enumerate(centroids_orig.index):
    ax.annotate(txt, (centroids_orig.iloc[txt]['y'] + annot_dx, centroids_orig.iloc[txt]['x']+ annot_dy), ha='center', color = 'white', fontsize = 15 )


plt.savefig('total_pass_clusters.png')
plt.show()
```

There is a risk here with so many clusters: overfitting. However, we want to ensure our model is as intepretable as possible, and having more clusters helps to visualize how teams are passing the ball.

What we can see here is that most of the clusters are significant predictors, and ending a pass outside of the dropped pass end 5 has a negative impact only compared to area 8. This is likely because its where the opponents goal is.

Any positive (and significant) coefficients increase the probability of a pass occuring, and vice versa.

I use a second logisitc regression to build a confusion matrix.

### Limitations

- Ranking top passers like this assumes risk is good - No concept of value. This analysis assumes that riskiness is good when overcome, however there is the possibility that players can take too much risk and not provide enough value. Without data on if passes led to shots, or goals, this is difficult to calculate.

- Clustering player types based solely on their most popular pass choice - over-simplification of the game. Many players change position throughought the game and actually fulfill multiple roles, especially those in midfield. Instead of comparing just the most popular cluster between players, a more rigorous similarity score should be applied.

- Like all sports there is much more nuance than simply looking at league level data. What is risk for one player or team is not necessarily risky for another. Players, and teams all play differently, so perhaps building individual classification models at these lower levels would give a more precise value of expected passes.
